SECTION 1 Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
```

SECTION 2 RMSE Comparison

```python
rmse_results = pd.DataFrame({

    "Model":[
        "Item-CF Random",
        "SVD Random",
        "Item-CF Temporal",
        "SVD Temporal",
        "Item-CF Per User Temporal",
        "SVD Per User Temporal",
        "Tuned SVD",
        "SVD++"
    ],

    "RMSE":[
        1.0881,
        0.9855,
        1.1248,
        1.0342,
        1.1544,
        0.9735,
        0.9703,
        0.9788
    ]
})

rmse_results.sort_values(
    "RMSE"
)
```

SECTION 3 RMSE Visualization

```python
plt.figure(figsize=(10,5))

plt.bar(
    rmse_results["Model"],
    rmse_results["RMSE"]
)

plt.xticks(rotation=45)

plt.title(
    "RMSE Comparison Across Models"
)

plt.ylabel(
    "RMSE"
)

plt.tight_layout()

plt.show()
```

SECTION 4 Ranking Metrics

```python
ranking_results = pd.DataFrame({

    "Metric":[
        "Precision@10",
        "Recall@10",
        "MAP@10"
    ],

    "Score":[
        0.4367,
        0.8977,
        0.5971
    ]
})

ranking_results
```

SECTION 5 Ranking Visualization

```python
plt.figure(figsize=(6,4))

plt.bar(
    ranking_results["Metric"],
    ranking_results["Score"]
)

plt.title(
    "Ranking Evaluation Metrics"
)

plt.ylabel(
    "Score"
)

plt.show()
```

SECTION 6 Cold Start Results

```python
cold_start_results = pd.DataFrame({

    "Segment":[
        "Cold",
        "Light",
        "Medium",
        "Heavy"
    ],

    "Ratings":[
        116863,
        124001,
        26493,
        446
    ],

    "RMSE":[
        1.697143,
        1.737353,
        1.910147,
        3.210828
    ]
})

cold_start_results
```

```python
plt.figure(figsize=(7,4))

plt.bar(
    cold_start_results["Segment"],
    cold_start_results["RMSE"]
)

plt.title(
    "Cold Start Performance"
)

plt.ylabel(
    "RMSE"
)

plt.show()
```

SECTION 7 Best Model

```python
best_model = rmse_results.sort_values(
    "RMSE"
).iloc[0]

best_model
```

 Final Conclusions

1. Matrix factorization methods consistently outperformed neighborhood-based collaborative filtering.

2. Tuned SVD achieved the best predictive performance with an RMSE of 0.9703.

3. Random evaluation produced optimistic results, while temporal evaluation provided a more realistic estimate of future performance.

4. Cold-start users remain a significant challenge in recommendation systems.

5. Under stricter ranking evaluation (users with at least 10 test interactions), the recommender achieved:

   - Precision@10 = 0.4367
   - Recall@10 = 0.8977
   - MAP@10 = 0.5971

6. The project demonstrates the effectiveness of matrix factorization techniques for large-scale movie recommendation.
